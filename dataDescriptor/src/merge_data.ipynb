```python
def merge_csvs(directory, comparison_df):
    merged_df = comparison_df
    merged_df['statecounty'] = merged_df['statecounty'].str.split().str[0]
    merged_df = merged_df.drop_duplicates(subset='statecounty', keep='first')

    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory, filename)
            print(file_path)
            
            # Read the CSV file
            csv_df = pd.read_csv(file_path)
            print(csv_df['statecounty'])
            csv_df['statecounty'] = csv_df['statecounty'].str.split().str[0]
            csv_df = csv_df.drop_duplicates(subset='statecounty', keep='first')


            merged_df = pd.merge(merged_df, csv_df, on='statecounty', how='left')
            
    return merged_df
        
```

```python
merged_df = merge_csvs('./processed_datasets', sorted_statecounty_df)
merged_df
```
